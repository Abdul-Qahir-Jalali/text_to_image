```python
from diffusers import StableDiffusionPipeline
import torch

# Load model, removing the revision argument
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    # revision="fp16"  # Remove or change to a valid revision
).to("cuda")
```

```python
from IPython.display import display

# Define your prompt
prompt = "A fantasy landscape with mountains, rivers, and a glowing sunset"

# Generate image
image = pipe(prompt).images[0]

# Display image properly
display(image)
```

```python
import gradio as gr

# Define the function that generates images
def generate_image(prompt):
    # Generate image from the model (you already have this code)
    image = pipe(prompt).images[0]
    return image

# Create a Gradio interface
iface = gr.Interface(fn=generate_image,
                     inputs=gr.Textbox(label="Enter Prompt", placeholder="Describe the image..."),
                     outputs=gr.Image(label="Generated Image"),
                     title="Text-to-Image Generator by Qahir",
                     description="Enter a prompt, and the model will generate an image based on your description.")

# Launch the interface (this runs it in Colab for testing)
iface.launch(share=True)
```
